```python
import numpy as np
import pandas as pd

values = results['pred_values']

thresholds = pd.DataFrame(index = values.index)

for col in values.columns:

    if('_d' in col):

        for t in [round(t, 2) for t in np.arange(0.05, 1, 0.05)]:

            thresholds[col + '_' + str(t)] = (values[col] >= t).astype(int)

    else: 

        thresholds[col] = values[col]
```

```python
from models import get_discrete_metrics

metrics_index = ['TP', 'FP', 'FN', 'TN', 'precision', 'recall', 'f1']

metrics = pd.DataFrame(index = metrics_index)

for model_threshold in thresholds.columns:

        if(model_threshold not in ['readmit_30d', 'readmit_90d']):

            true_col = ('readmit_30d' if 'd30' in model_threshold else 'readmit_90d')

            metrics.loc['TP', model_threshold] = ((thresholds[model_threshold] == 1) & (thresholds[true_col] == 1)).sum()
            metrics.loc['FP', model_threshold] = ((thresholds[model_threshold] == 1) & (thresholds[true_col] == 0)).sum()
            metrics.loc['FN', model_threshold] = ((thresholds[model_threshold] == 0) & (thresholds[true_col] == 1)).sum()
            metrics.loc['TN', model_threshold] = ((thresholds[model_threshold] == 0) & (thresholds[true_col] == 0)).sum()

            y_true = thresholds[true_col].astype(int)

            metrics = get_discrete_metrics(y_true, thresholds[model_threshold], model_threshold, metrics, transposed = True)
```

```python
from cost import separate_model_threshold

def estimate_intervention_cost(row, df_pred, df_cost, model, threshold):

    relative_prob = df_pred.loc[row, model] / threshold

    extra_day_stay_cost = np.nanmin([df_cost.loc[row, 'cost_per_day_stay'], df_cost.loc[row, 'avg_cost_of_prev_stays']])

    intervention_cost = np.min([np.floor(relative_prob), 3]) * extra_day_stay_cost

    return intervention_cost

d30 = True

flag = ('d30' if d30 else 'd90')

gains = pd.DataFrame(index = df_cost.index)

for col in thresholds.columns:

    if (flag in col):

        model, threshold = separate_model_threshold(col)

        for row in thresholds.index:

            if(thresholds.loc[row, col] == 1):

                exp_avoided_cost = 0.2 * results['pred_values'].loc[row, model] * df_cost.loc[row, 'total_readmission_cost']

                intervention_cost = estimate_intervention_cost(row, results['pred_values'], df_cost, model, threshold)

                gains.loc[row, col] = exp_avoided_cost - intervention_cost

            else: 

                gains.loc[row, col] = 0
                
gains.loc['total_avoided'] = gains.sum(axis = 0)
```
