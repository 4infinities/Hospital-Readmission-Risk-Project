```python
avoided.drop(columns = ['readmit_30d', 'readmit_90d'], inplace = True)
avoided.set_index('Unnamed: 0', inplace = True)
avoided
```

```python
pct_avoided = avoided.copy()
total_readmit_30d = df_cost[df_cost["readmit_30d"] == 1]["total_readmission_cost"].sum()
total_readmit_90d = df_cost[df_cost["readmit_90d"] == 1]["total_readmission_cost"].sum()

for col_name, col in avoided.items():
    pct_avoided[col_name] = col/(total_readmit_30d if "_d30" in col_name else total_readmit_90d)
```

```python
print(pct_avoided.max(axis = 1))
```

```python
model_scoring = pd.Series(index = pct_avoided.columns)
max_values = set(pct_avoided.max(axis = 1))

for model, pct_values in pct_avoided.items():

    values = set(pct_values)
    model_scoring[model] = len(values.intersection(max_values))

model_scoring.sort_values(ascending = False)
```
