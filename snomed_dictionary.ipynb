```python
import json 

BASE_URL = "https://snowstorm.snomedtools.org/fhir" 
SNOMED_URL = "http://snomed.info/sct"

def snomed_lookup(code: str):
    """Call FHIR CodeSystem/$lookup for a SNOMED CT code and return JSON."""
    params = {
        "system": SNOMED_URL,
        "code": code
    }
    url = f"{BASE_URL}/CodeSystem/$lookup"
    resp = requests.get(url, params=params)
    resp.raise_for_status()
    return resp.json()

result = snomed_lookup("22298006")
print(json.dumps(result, indent=2)[:2000])
```
